```python
import pandas as pd

df = pd.read_csv('/content/Iris.csv')

print("Dataset loaded successfully. Displaying the first 5 rows:")
df.head()
```

```python
print("\n--- DataFrame Information ---")
df.info()

print("\n--- Missing Values ---")
print(df.isnull().sum())

print("\n--- Descriptive Statistics ---")
df.describe()
```

**Reasoning**:
I will first remove any duplicate rows and reset the index, then drop the 'Id' column, and finally rename the specified columns to improve readability and consistency as per the instructions. I will display the head of the DataFrame to show the changes.



```python
print("Checking for and removing duplicate rows...")
initial_rows = df.shape[0]
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
duplicates_removed = initial_rows - df.shape[0]
print(f"Removed {duplicates_removed} duplicate rows.")

print("\nDropping 'Id' column...")
df.drop('Id', axis=1, inplace=True)
print("'Id' column dropped.")

print("\nRenaming columns for better readability...")
df.rename(columns={
    'SepalLengthCm': 'sepal_length',
    'SepalWidthCm': 'sepal_width',
    'PetalLengthCm': 'petal_length',
    'PetalWidthCm': 'petal_width',
    'Species': 'species'
}, inplace=True)
print("Columns renamed.")

print("\nCleaned DataFrame head:")
print(df.head())

print("\nCleaned DataFrame Info:")
df.info()
```

```python
print("Creating new features: 'sepal_area' and 'petal_area'...")
df['sepal_area'] = df['sepal_length'] * df['sepal_width']
df['petal_area'] = df['petal_length'] * df['petal_width']
print("New features created.")

print("\nDataFrame head with new features:")
print(df.head())

print("\nDataFrame Info with new features:")
df.info()
```

```python
output_file_path = 'Iris_cleaned.csv'
df.to_csv(output_file_path, index=False)
print(f"Cleaned data saved successfully to '{output_file_path}'")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 10))

# 1. Scatter plot of sepal_length vs. sepal_width
plt.subplot(2, 2, 1) # 2 rows, 2 columns, 1st plot
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', palette='viridis')
plt.title('Sepal Length vs. Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')

# 2. Scatter plot of petal_length vs. petal_width
plt.subplot(2, 2, 2) # 2 rows, 2 columns, 2nd plot
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', palette='viridis')
plt.title('Petal Length vs. Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')

# 3. Distribution plot for sepal_area
plt.subplot(2, 2, 3) # 2 rows, 2 columns, 3rd plot
sns.histplot(data=df, x='sepal_area', hue='species', kde=True, palette='viridis')
plt.title('Distribution of Sepal Area by Species')
plt.xlabel('Sepal Area (cm^2)')
plt.ylabel('Frequency')
plt.legend(title='Species', labels=df['species'].unique())

# 4. Distribution plot for petal_area
plt.subplot(2, 2, 4) # 2 rows, 2 columns, 4th plot
sns.histplot(data=df, x='petal_area', hue='species', kde=True, palette='viridis')
plt.title('Distribution of Petal Area by Species')
plt.xlabel('Petal Area (cm^2)')
plt.ylabel('Frequency')
plt.legend(title='Species', labels=df['species'].unique())

plt.tight_layout()
plt.show()
```
